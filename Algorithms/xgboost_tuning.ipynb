```python
import pandas as pd
import csv
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from ipynb.fs.full.set_up import data_prep
```

```python


X_train_scaled,  X_test_scaled  ,y_train, y_test , X_train, X_test = data_prep()


pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', XGBRegressor())
])


# Fit the model
pipeline.fit(X_train_scaled, y_train)


# Define grid search parameters
param_grid = {
    'model__learning_rate': [0.07, 0.1, 0.15],
    'model__max_depth': [5, 7, 10],
    'model__n_estimators': [150, 200, 250],
}

# Perform grid search
# n_jobs = -1 ----> use all processors (Number of jobs to run in parallel)
# cv ----> number of cross-validation (Tests)
grid_search = GridSearchCV(pipeline, param_grid=param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)


res = pd.DataFrame(grid_search.cv_results_)

print("-------")
print(grid_search.best_index_)
print("-------")  
# Print best parameters
print(grid_search.best_params_)

# Re-fit model with best parameters
best_model = grid_search.best_estimator_
best_model.fit(X_train, y_train)

# Predict on test set
y_pred = best_model.predict(X_test)

# Evaluate model
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('MAE:', mae)
print('R2:', r2)
res.sort_values('rank_test_score', inplace = True)
# {'model__learning_rate': 0.07, 'model__max_depth': 7, 'model__n_estimators': 250}
# MAE: 386721.6722941933
# R2: 0.793217263296604
```
