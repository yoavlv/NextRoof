```python
def create_connection(url,headers=None ,cookies =None , params = None , json = False):
    retries = Retry(total=2, backoff_factor=0.5, status_forcelist=[500, 502, 503, 504])
    adapter = HTTPAdapter(max_retries=3)
    session = requests.Session()
    session.mount('https://', adapter)
    session.mount('http://', adapter)
    response = session.get(url, params=params, headers=headers, cookies=cookies)
    
    if response.status_code != 200:
        print(f"Error (create_connection) status_code {response.status_code}")
        return False
    
    if 'ShieldSquare Captcha' in response.text:
        print(f"(create_connection) ShieldSquare Captcha ")
        return False
        
    if json:
        return response.json()
    
    # soup content
    soup = BeautifulSoup(response.content, 'html5lib')
    return soup

url = 'https://gisn.tel-aviv.gov.il/arcgis/rest/services/IView2/MapServer/513/query?where=1%3D1&outFields=*&f=json'
url = 'https://gisn.tel-aviv.gov.il/arcgis/rest/services/IView2/MapServer/513/query?where=1%3D1&outFields=*&resultRecordCount=1000&f=json'


```
